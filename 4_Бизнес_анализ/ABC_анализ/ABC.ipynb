# ABC - анализ

- показывает какой товар приносит максимум денег
- основан на правиле Парето

## Одномерный

```python
# получаем количество каждого товара
grouped_df = df.groupby('DR_NDrugs').agg({'DR_Kol': sum})
```

```python
# получаем вклад каждого проданного товара в общее количество
grouped_df = (grouped_df['DR_Kol'] / sum(grouped_df['DR_Kol'])).sort_values(ascending=False)
```

```python
# вклад накопительным итогом
grouped_df = grouped_df.cumsum().reset_index()
```

```python
grouped_df.head()
```

```python
grouped_df['abc'] = np.where(grouped_df['DR_Kol'] < 0.8, 'A', np.where(grouped_df['DR_Kol'] < 0.95, 'B', 'C'))
grouped_df.tail()
```

## Многомерный

```python
df = pd.read_csv("../data_drugs.csv", encoding='1251')
df.head()
```

```python
# получаем количество каждого товара и суммарную стоимость по каждому товару
grouped_df = df.groupby('DR_NDrugs').agg({'DR_Kol': sum, 'DR_CRoz': sum})
grouped_df.head(3)
```

```python
grouped_df['rel_kol'] = (grouped_df['DR_Kol'] / sum(grouped_df['DR_Kol']))
grouped_df.sort_values(by='rel_kol', ascending=False, inplace=True)
grouped_df['cumsum_kol'] = grouped_df['rel_kol'].cumsum()
grouped_df['abc_kol'] = np.where(grouped_df['cumsum_kol'] < 0.8, 'A', np.where(grouped_df['cumsum_kol'] < 0.95, 'B', 'C'))
grouped_df.tail()
```

```python
grouped_df['rel_roz'] = (grouped_df['DR_CRoz'] / sum(grouped_df['DR_CRoz']))
grouped_df.sort_values(by='rel_roz', ascending=False, inplace=True)
grouped_df['cumsum_roz'] = grouped_df['rel_roz'].cumsum()
grouped_df['abc_roz'] = np.where(grouped_df['cumsum_roz'] < 0.8, 'A', np.where(grouped_df['cumsum_roz'] < 0.95, 'B', 'C'))
grouped_df
```
