# Теория

## Графы

## Дерево. Лес

- деревом называется связанный грраф, не содержащий циклов. все вершины соединены единственным путем.
- Дерево: N вершин, N - 1 ребро
- лесом называется граф, состоящеий из компонент дерева
- корнем дерева является главная вершина дерева
- листья - элементы дерева, не имеющие дочерних элементов
- узел - элемент дерева, имеющий потомков
- потомок - вершина, находящаяся ниже другой и имеющая с ней прямую связь
- предок - вершина, находящаяся выше другой и имеющая с ней прямую связь

## Способы хранения графа

- матрица смежности
- список смежности
- список ребер

### Матрица смежности

- наиболее эффективный по времени проверки наличия ребра способ хранения графа
- если граф взвешенный то в ячейке [i][j] будет храниться число, равное весу ребра
