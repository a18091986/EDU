# Теория

```python
- инцидентность ребра вершине говорит о том, что ребро выходит из данной вершины
- степень вершины - количество ребер выходящих из данной вершины (инцидентным ей)
- кратные ребра - ребра, соединяющие одинаковые пары вершин (т.е. инцидентные одинаковым парам вершин)
- смежные вершины - соединенные хотя бы одним ребром
- смежные ребра - имеющие хотя бы одну общую вершину
- изолиорованная вершина - не имеющая ребер
- висячая вершина - имеющая ровно одно ребро
- петля - ребро, выходящее и возвращающееся в одну и ту же вершину
- если в графе есть петли, его называют псевдографом
- обыкновенный граф - граф без петель и кратных ребер
- степень вершины с петлей = 2
- полный граф - граф, в котором есть ребро между каждой парой вершин
- связный граф - граф, у которого есть цепь между каждой парой вершин (нет изолированных вершин)
- путь - последовательность вершин и ребер, в которой каждый элемент кроме начальной и конечной вершины инцидентен предыдущему и последующему
- цикл - путь, у которого начальниая и конечная вершины совпадают
- вершинно-простой путь - путь, не проходящий ни через одну вершину дважды
- реберно-простой путь - путь, не проходящий ни через одно ребро дважды
```
