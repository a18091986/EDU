Описание данных

Имеются следующие переменные:

- Hotel – тип отеля (City Hotel или Resort Hotel)  
- Is canceled – бронирование было отменено (1) или нет (0); неотменённое считается успешным
- Lead time – количество дней, прошедших между датой бронирования и датой прибытия  
- Arrival full date – полная дата прибытия
- Arrival date year – год прибытия  
- Arrival date month – месяц прибытия  
- Arrival date week number – номер недели прибытия
- Arrival date day of month – день прибытия
- Stays in weekend nights – количество выходных (суббота или воскресенье), которые гость забронировал для проживания в отеле
- Stays in week nights – количество дней (с понедельника по пятницу), которые гость забронировал для проживания в отеле
- Stays total nights – общее число забронированных ночей (сумма двух предыдущих колонок)
- Adults – число взрослых
- Children – число детей
- Babies – число младенцев 
- Meal – выбранный тип питания
- Country – страна происхождения клиента
- Reserved room type – тип зарезервированного номера
- Assigned room type – тип полученного номера (может отличаться от забронированного)
- Customer type – тип бронирования
- Reservation status – значение последнего статуса брони: Canceled – было отменено клиентом; Check-Out – клиент зарегистрировался, но уже покинул отель; No-Show – клиент не зарегистрировался и сообщил администрации отеля причину
- Reservation status date – дата обновления статуса

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 40)
```

```python
bookings = pd.read_csv('https://stepik.org/media/attachments/lesson/360344/bookings.csv', sep=';', 
                       parse_dates=['arrival full date', 'Reservation status_date'])
bookings.head(3)
```

```python
# Проверьте размер таблицы, типы переменных, а затем выведите первые 7 строк, чтобы посмотреть на данные. 
print(bookings.dtypes)
print(bookings.shape) 
bookings.head(7)
```

```python
# Приведите названия колонок к нижнему регистру и замените пробелы на знак нижнего подчеркивания.
new_columns_names = [x.lower().replace(' ', '_') for x in bookings.columns]
bookings = bookings.rename(columns = dict(zip(bookings.columns, new_columns_names)))
bookings.head(1)
```

```python
# Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.
pd.pivot_table(bookings.loc[bookings['is_canceled'] == 0], index='country', values=['is_canceled'], aggfunc='count').sort_values(by='is_canceled', ascending=False).head()
```

```python
# На сколько ночей (stays_total_nights)  в среднем бронируют отели типа 
# City Hotel? Resort Hotel? Запишите полученные значения в пропуски с точностью до 2 знаков после точки.
pd.pivot_table(bookings, index='hotel', values=['stays_total_nights'], aggfunc='mean').round(2)
```

```python
# Иногда тип номера, полученного клиентом (assigned_room_type), 
# отличается от изначально забронированного (reserved_room_type). 
# Такое может произойти, например, по причине овербукинга.

#Сколько подобных наблюдений встретилось в датасете?

#*отмена бронирования также считается
bookings.loc[bookings['assigned_room_type'] != bookings['reserved_room_type']].count().head(1)
```

```python
# На какой месяц чаще всего оформляли бронь в 2016 году? Изменился ли самый популярный месяц в 2017 году?
pd.pivot_table(bookings, index=['arrival_date_year'], columns=['arrival_date_month'], values=['hotel'], aggfunc='count')
bookings.groupby(['arrival_date_year', 'arrival_date_month'])['hotel'].agg('count').sort_values()
bookings.groupby(['arrival_date_year']).arrival_date_month.agg(pd.Series.mode)
bookings.groupby(['arrival_date_year', 'arrival_date_month'])['hotel'].agg('count').groupby(['arrival_date_year']).nlargest(1)
```

```python
#Сгруппируйте данные по годам, а затем проверьте, на какой месяц (arrival_date_month) 
#бронирования отеля типа City Hotel отменялись чаще всего в 2015? 2016? 2017? 
#Hint: для этого сначала примените метод groupby, после чего выберите нужную колонку для дальнейших действий :) Например:
# df.groupby('column_1')['column_2'].what_to_do()

bookings.loc[(bookings['is_canceled'] == 1) & (bookings['hotel'] == 'City Hotel')]\
        .groupby('arrival_date_year').arrival_date_month.agg(pd.Series.mode)
```

```python
# Посмотрите на числовые характеристики трёх колонок: adults, children и babies. 
# Какая из них имеет наибольшее среднее значение?
bookings.adults.mean(), bookings.children.mean(), bookings.babies.mean()
bookings[['adults', 'children', 'babies']].describe()
```

```python
#Создайте колонку total_kids, объединив столбцы children и babies. 
#Для отелей какого типа среднее значение переменной оказалось наибольшим?
#City hotel – отель находится в городе
#Resort hotel – отель курортный
#В качестве ответа укажите наибольшее среднее total_kids, округлив до 2 знаков после точки.

bookings.assign(total_kids=bookings.children + bookings.babies)
bookings = bookings.eval('total_kids = children + babies')
bookings.groupby('hotel')['total_kids'].agg('mean').round(2)
```

```python
# Не все бронирования завершились успешно (is_canceled), поэтому можно посчитать, 
# сколько клиентов было потеряно в процессе. Иными словами, посчитать метрику под названием Churn Rate.

bookings['has_kids'] = np.where(bookings['total_kids'] > 0, 0, 1)
kids_cr = bookings.loc[(bookings['has_kids'] == 1) & (bookings['is_canceled'] == 1)].count()[0] / bookings.loc[(bookings['has_kids'] == 1)].count()[0]
no_kids_cr = bookings.loc[(bookings['has_kids'] == 0) & (bookings['is_canceled'] == 1)].count()[0] / bookings.loc[(bookings['has_kids'] == 0)].count()[0] 
kids_cr, no_kids_cr
# bookings.pivot_table(index='is_canceled',  values=['has_kids'], aggfunc='sum')
```
